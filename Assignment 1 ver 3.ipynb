```python
#Check for missing data
def missing(df):
    total = df.isnull().sum().sort_values(ascending=False)
    percent =(df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missingdata = pd.concat([total, percent], axis=1, keys=["Total", "Percent"])
    return missingdata
```

```python
# We can identify categorical features in data set by number of unique classes in their respective columns
# Categorical features in this set are Fylke, Område, Kommune, Linjetype, Linjefylke
# Encoding categorical features:
```

```python
#Fylke
fylke_encod = {"Viken" : 1, "Oslo" : 2}

df["Fylke"] = df["Fylke"].map(fylke_encod)
df["Fylke"].value_counts()
```

```python
#Område
omrade_encod = {"Vest" : 1, "Nordøst" : 2, "Indre By" : 3, "Sør" : 4}

df["Område"] = df["Område"].map(omrade_encod)
df["Område"].value_counts()
```

```python
#Linjetype
linjetype_encod = {"Lokal" : 1, "Skole" : 2, "Ekspress" : 3, "Natt" : 4}

df["Linjetype"] = df["Linjetype"].map(linjetype_encod)
df["Linjetype"].value_counts()
```

```python
#Linjefylke
linjefylke_encod = {"Viken" : 1, "Oslo" : 2}

df["Linjefylke"] = df["Linjefylke"].map(linjefylke_encod)
df["Linjefylke"].value_counts()
```

```python
#Kommune
cat_columns1 = ["Kommune"]
encod = ce.OrdinalEncoder(cols=cat_columns1)
df = encod.fit_transform(df)
```

```python
#Turid column is unnecessary in further analysis
df.drop("TurId", axis=1, inplace=True)
```

```python
#Name of some columns are confusing to read. Adjusting naming
df.rename(columns={"Tidspunkt_Faktisk_Ankomst_Holdeplass_Fra" : "Actual_arrival",
                  "Tidspunkt_Faktisk_Avgang_Holdeplass_Fra" : "Actual_departure",
                  "Tidspunkt_Planlagt_Ankomst_Holdeplass_Fra" : "Planned_arrival",
                  "Tidspunkt_Planlagt_Avgang_Holdeplass_Fra" : "Planned_departure",
                  }, inplace=True)
```

```python
# We can identify 5 date features out of which 4 are in time format
# Date features in this set are: Dato, Actual_arrival_bus_stop, Actual_departure_bus_stop, Planned_arrival_bus_stop, Planned_departure_bus_stop
```

```python
df["Actual_arrival"] = df["Actual_arrival"].astype(str).str.replace("24:", "00:")
df["Actual_arrival"] = df["Actual_arrival"].astype(str).str.replace("25:", "00:")

df["Actual_departure"] = df["Actual_departure"].astype(str).str.replace("24:", "00:")
df["Actual_departure"] = df["Actual_departure"].astype(str).str.replace("25:", "00:")

df["Planned_arrival"] = df["Planned_arrival"].astype(str).str.replace("24:", "00:")
df["Planned_arrival"] = df["Planned_arrival"].astype(str).str.replace("25:", "00:")

df["Planned_departure"] = df["Planned_departure"].astype(str).str.replace("24:", "00:")
df["Planned_departure"] = df["Planned_departure"].astype(str).str.replace("25:", "00:")
```
